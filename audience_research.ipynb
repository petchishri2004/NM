```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = {
    'Age': [10, 15, 20, 25, 30, 35, 40, 45, 50, 55],
    'Gender': ['Female', 'Male', 'Female', 'Male', 'Female', 'Female', 'Female', 'Female', 'Male', 'Female'],
    'Income': [30000, 40000, 50000, 60000, 70000, 80000, 90000, 100000, 110000, 120000],
    'Education': ['Bachelor', 'PhD', 'PhD', 'PhD', 'Master', 'PhD', 'Bachelor', 'Bachelor', 'PhD', 'Bachelor'],
    'Viewership': [500000, 1000000, 1500000, 2000000, 2500000, 3000000, 3500000, 4000000, 4500000, 5000000]
}
df = pd.DataFrame(data)
print(df.head())
print(df.describe())
sns.histplot(df['Age'], bins=5)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
gender_counts = df['Gender'].value_counts()
gender_counts.plot(kind='bar')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
sns.boxplot(x=df['Income'])
plt.title('Income Distribution')
plt.xlabel('Income')
plt.show()
education_counts = df['Education'].value_counts()
education_counts.plot(kind='barh')
plt.title('Education Level Distribution')
plt.xlabel('Count')
plt.ylabel('Education Level')
plt.show()
sns.scatterplot(data=df, x='Age', y='Viewership', hue='Gender')
plt.title('Viewership vs. Age')
plt.xlabel('Age')
plt.ylabel('Viewership')
plt.legend(title='Gender')
plt.show()


```
